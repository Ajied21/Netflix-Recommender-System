```python
import pandas as pd
import numpy as np
import sys
import json
pd.set_option('display.max_columns', 999)

data = pd.read_csv('/home/ajied/FInal_Project/Streamlit_App/Data/Netflix(Movie & Show).csv')
data.head()
```

```python
# Create a new data frame with relevant columns only
df_all = data[['title', 'description', 'type','release_year', 'genres', 'Poster', 'Netflix Link', 'TMDb Trailer', 'imdb_score',]].copy()
df_all.head()
```

```python
# Create a new data frame with relevant columns only
df = data[['title', 'description', 'type','release_year', 'genres', 'Poster', 'Netflix Link', 'TMDb Trailer', 'imdb_score',]].copy()
df = df[(df['type']=='MOVIE')]
df.head()
```

```python
# Create a new data frame with relevant columns only
df_tv = data[['title', 'description', 'type','release_year', 'genres', 'Poster', 'Netflix Link', 'TMDb Trailer', 'imdb_score',]].copy()
df_tv = df_tv[(df_tv['type']=='SHOW')]
df_tv.head()
```

```python
np.set_printoptions(threshold=sys.maxsize)
# Fetch genres of all movies
genres_alls = [df_all.loc[i]['genres'].split(', ') for i in df_all.index]
# Find the list of genres of all movies in alphabetical order
genres_all = sorted(list(set([item for sublist in genres_alls for item in sublist])))
genres_all
```

```python
# Fetch genres of all movies
genres_all_movies = [df.loc[i]['genres'].split(', ') for i in df.index]
# Find the list of genres of all movies in alphabetical order
genres = sorted(list(set([item for sublist in genres_all_movies for item in sublist])))
genres
```

```python
# Fetch genres of all movies
genres_all_tv = [df_tv.loc[i]['genres'].split(', ') for i in df_tv.index]
# Find the list of genres of all movies in alphabetical order
genres_tv = sorted(list(set([item for sublist in genres_all_tv for item in sublist])))
genres_tv
```

```python
print(f'genre movie & tv : {len(genres_all)}\n')
print(f'genre movie : {len(genres)}\n')
print(f'genre tv : {len(genres_tv)}')
```

```python
# Initialize lists for movie data as well as titles
full_movie = []
movie_titles = []

full_tv = []
tv_titles = []
```

```python
print(df,'\n')
print(df_tv)
```

```python
for i in df.index:
    # Append movie title and the index of the movie
    movie_titles.append((df.loc[i]['title'].strip(), i, df.loc[i]['Netflix Link'].strip()))
    # Add list of genres of the movies (1/0) to movie data
    movie_data = [1 if genre in df.loc[i]['genres'].split(',') else 0 for genre in genres]
    # Add IMDb score of the movie to the movie data
    movie_data.append(df.loc[i]['imdb_score'])
    # Add record of movie to main data list
    full_movie.append(movie_data)
```

```python
for z in df_tv.index:
    # Append movie title and the index of the movie
    tv_titles.append((df_tv.loc[z]['title'].strip(), z, df_tv.loc[z]['Netflix Link'].strip()))
    # Add list of genres of the movies (1/0) to movie data
    tv_data = [1 if genre in df_tv.loc[z]['genres'].split(',') else 0 for genre in genres]
    # Add IMDb score of the movie to the movie data
    tv_data.append(df_tv.loc[z]['imdb_score'])
    # Add record of movie to main data list
    full_tv.append(tv_data)
```

```python
print(movie_titles[1],'\n')
print(tv_titles[1])
```
